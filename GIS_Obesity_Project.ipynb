```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/GISProject_Obesity/analytic_data2023_0.csv')
```

```python
df.columns.tolist()
```

```python
column_names = ['State FIPS Code',
 'County FIPS Code',
 '5-digit FIPS Code',
 'State Abbreviation',
 'Name','Adult Obesity raw value'] #list of desired column names
```

```python
newdf = df[column_names]
```

```python
newdf = newdf.iloc[1:]
#removes first row - described the data
```

```python
newdf["Adult Obesity raw value"] = pd.to_numeric(newdf["Adult Obesity raw value"])
#ensures everything is numberic
```

```python
newdf_sorted = newdf.sort_values(by="Adult Obesity raw value", ascending=True)
#sorts by increasing obesity by county
```
